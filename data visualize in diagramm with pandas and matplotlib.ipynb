```python
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib inline

dict = {"Year": [2010,2010,2011,2011,2012,2012,2013,2013,2014,2014,2015,2015],"Sales":[310,290,295,295,325,299,330,300,335,301,340,302],"place":["home","office","office","home","home","office","home","office","home","office","home","office"]}
df = pd.DataFrame(dict) # converts a dictionary of sales data in a DataFrame
df_home = df[df["place"]=="home"] # DataFrame of sales data at home
df_office = df[df["place"]=="office"]

fig, ax = plt.subplots()
# visualize the sales data at home and in the office
df_home.plot(x="Year",y="Sales",ax=ax)
df_office.plot(x="Year",y="Sales",ax=ax)
ax.legend(["home","office"])
plt.show()
```

```python
dict2 ={"Year": [2011,2012,2013,2014,2015,2016],"Sales":[310,290,325,330,335,340]}
df_sales = pd.DataFrame(dict2)
year = df_sales["Year"]
sales = df_sales["Sales"]

# visualize the datd of sales in diagramm
plt.plot(year,sales,label = "Sales")
plt.legend()
plt.show()
```
